```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
header_row = ['age','sex','cp','BP','chol','fbs','ecg','maxhr','eiang','eist','slope','vessels','thal','diagnosis']
df = pd.read_excel("C:/Users/carllel colquhoun/Desktop/PYTHON FILES/LabWeek5/CardiologyNumerical.xls",names=header_row)
```

```python
df = df.drop([0,1],axis=0)
```

```python
#K-means clustering is a clustering algorithm that aims to partition n(302) observations into k clusters
```

```python
X = df[['maxhr','BP']].values
```

```python
#create DF
#x = df[['BP']].values
#y=df[['chol']].values
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
#The chart below shows the dataset for 302 patients, with the Cholesterol feature 
#on the x-axis and High Blood Pressure feature on the y-axis.
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='rainbow')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.title('High Blood Pressure vs Maximum Heart Rate')
plt.xlabel('Maximum Heart Rate')
plt.ylabel('High Blood Pressure')
```
